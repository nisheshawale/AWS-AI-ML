### Importing necessary libraries

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
```

### Import custom made data from S3 bucket

```python
bucket='dataforai'
data_key = 'criminal_profile_data.xlsx'
data_location = 's3://{}/{}'.format(bucket, data_key)

df = pd.read_excel(data_location)
df
```

### Using one hot encoding for the categorical data

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
df.shape
```

```python
df.head()
```

### Taking one data from each class(criminal and non criminal) to predict the value later. Train test split is not used as the size of the dataset is small.

```python
test_1 = df[0:1]
test_2 = df[24:25]
df = df.drop([df.index[0], df.index[24]]) 
df
```

### Shuffling the data before separating the label from the dataframe

```python
df = df.sample(frac = 1)
df
```

### Separating the features and label to provide input for kNN.

```python
Y = df['Is Criminal']
X = df.drop(['Is Criminal'], axis = 1)
print(X.shape)
print(Y.shape)
```

### Training k nearest neighbor classifier on the data

```python
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X, Y)
```

### The label is dropped from the test_1 and test_2 to pass them as an input to knn for prediction.

```python
pred_criminal = test_1.drop(['Is Criminal'], axis = 1) 
pred_non_criminal = test_2.drop(['Is Criminal'], axis = 1) 
```

```python
map_values = {1: 'Criminal', 0: 'not Criminal'}
```

# Checking the prediction of the model from the data which was separated above for testing.

### Prediction for the criminal profile

```python
predicted_1 = map_values[np.asscalar(neigh.predict(pred_criminal))]
print('The given profile suggests that the person is {}.'.format(predicted_1))
```

### Prediction for the non criminal profile

```python
predicted_2 = map_values[np.asscalar(neigh.predict(pred_non_criminal))]
print('The given profile suggests that the person is {}.'.format(predicted_2))
```
